### Embedding dan Persiapan Akses OPENAI API
Menggunakan OPENAI Embedding untuk melakukan embed terhadap dokumen input untuk kemudian disimpan pada database vector ChromaDB

```python
# load OPENAI API Key dari .env
from langchain_openai import OpenAIEmbeddings
from dotenv import load_dotenv,find_dotenv
print("Mengambil OPENAI API key? ", load_dotenv(find_dotenv()), "\n")

import openai
from openai import OpenAI as llm

# embedding
embedding_function = OpenAIEmbeddings()

# access OPENAI API
OPENAI_API_KEY = os.getenv("OPENAI_API_KEY")
client = llm(api_key=OPENAI_API_KEY)
```

```python
texts = ["Pasal 6 merujuk kepada UU KUP", "Pasal 6 merujuk kepada UU Nomor 6 Tahun 1983"]

for text in texts:
  print("Original text: ", text)
  print("Contextual Reps: ", embedding_function.embed_query(text))



# def get_embedding(text, model="text-embedding-3-small"):
#    text = text.replace("\n", " ")
#    return client.embeddings.create(input = [text], model=model).data[0].embedding
```
